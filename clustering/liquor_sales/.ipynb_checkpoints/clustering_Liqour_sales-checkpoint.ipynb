```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.preprocessing import normalize
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
```

```python
location = pd.read_csv('C:/Users/allan/src/projects/clustering/liquor_sales.csv', low_memory=False)

location.info()
```

```python
location_var = location
location_var = location_var.drop(columns=['invoice_and_item_number','address', 'store_name','date', 'county','category_name', 'vendor_number', 'vendor_name', 'item_description','sale_dollars'])
location_var.head()
```

```python
lb = LabelEncoder()
location_var['city_en'] = lb.fit_transform(location_var['city'])
location_var = location_var.drop(columns=['city'])

location_var.head()
```

```python
location_var[['longitude', 'latitude']] = location['store_location'].str.extract(r'POINT \((-?\d+\.\d+) (-?\d+\.\d+)\)')

location_var['longitude'] = location_var['longitude'].astype(float)
location_var['latitude'] = location_var['latitude'].astype(float)

sns.scatterplot(data = location_var, x= location_var['longitude'], y = location_var['latitude'], hue = 'city_en')
plt.xlabel('longitude')
plt.ylabel('latitude')

plt.show()
```

```python
location_var.dropna(subset=['longitude', 'latitude', 'city_en'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(location_var[['longitude', 'latitude']] , location_var['city_en'], test_size=0.33, random_state=0)
```

```python
X_train_norm = preprocessing.normalize(X_train)
X_test_norm = preprocessing.normalize(X_test)
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 0, n_init='auto')
kmeans.fit(X_train_norm)

all_data_norm = normalize(location_var[['longitude', 'latitude']])
cluster = kmeans.predict(all_data_norm)

location_var['cluster'] = cluster

if len(location) == len(location_var):
    location['clusters'] = location_var['cluster']
else:
    location = location.merge(location_var[['cluster']], left_index=True, right_index=True, how='left')
    
#location['clusters'] = location_var['cluster']
location.head()
```

```python
train_clusters = kmeans.predict(X_train_norm)
sns.scatterplot(data=X_train, x='longitude', y='latitude', hue=train_clusters)

plt.xlabel('longitude')
plt.ylabel('latitude')

plt.show()
```

```python
location['cluster'] = location['cluster'].astype('category')
location = location.drop_duplicates()
location.head()
location.to_csv('C:/Users/allan/src/projects/clustering/liquor_sales_clustering.csv', index=False)
```
