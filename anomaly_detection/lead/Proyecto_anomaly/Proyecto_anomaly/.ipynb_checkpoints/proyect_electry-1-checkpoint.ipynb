```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python

file_path = 'C:/Users/Charly/Desktop/Proyecto_anomaly/test.csv'
data = pd.read_csv(file_path)

data_head = data.head()
data_info = data.info()
data_description = data.describe()

data_head, data_info, data_description
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'], errors='coerce')

missing_values = data.isnull().sum()
print("Valores faltantes:\n", missing_values)

meter_reading_stats = data['meter_reading'].describe()
print("\nEstadísticas de 'meter_reading':\n", meter_reading_stats)
```

```python
data['meter_reading'].fillna(data['meter_reading'].mean(), inplace=True)

meter_reading_stats = data['meter_reading'].describe()
print("Estadísticas de 'meter_reading':\n", meter_reading_stats)
```

```python
data['year'] = data['timestamp'].dt.year
data['month'] = data['timestamp'].dt.month
data['hour'] = data['timestamp'].dt.hour

# Agrupar el consumo promedio por día
daily_consumption = data.groupby(data['timestamp'].dt.date)['meter_reading'].mean()
print("Consumo diario promedio:\n", daily_consumption.head())
```

```python

data_for_model = data[['meter_reading', 'month', 'hour']]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_for_model)
```

```python
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # 1% de datos esperados como anomalías
iso_forest.fit(data_scaled)
data['anomaly'] = iso_forest.predict(data_scaled)
data['anomaly'] = data['anomaly'].apply(lambda x: 1 if x == -1 else 0)


```

```python
plt.figure(figsize=(14, 7))
plt.plot(data['timestamp'], data['meter_reading'], label='Consumo de energía')
plt.scatter(data[data['anomaly'] == 1]['timestamp'], 
            data[data['anomaly'] == 1]['meter_reading'], 
            color='red', label='Anomalía', marker='o')
plt.xlabel('Fecha')
plt.ylabel('Consumo de Energía')
plt.legend()
plt.show()  
```

```python
#IQR
Q1 = data['meter_reading'].quantile(0.25)
Q3 = data['meter_reading'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Detectar anomalías
outliers_iqr = data[(data['meter_reading'] < lower_limit) | (data['meter_reading'] > upper_limit)]

# Análisis con desviación estándar
mean = data['meter_reading'].mean()
std_dev = data['meter_reading'].std()

threshold = 3
outliers_std = data[(data['meter_reading'] > mean + threshold * std_dev) | 
                     (data['meter_reading'] < mean - threshold * std_dev)]

print(f"Anomalías por IQR: {outliers_iqr.shape[0]} registros")
print(f"Anomalías por desviación estándar: {outliers_std.shape[0]} registros")
```

```python
data_for_model = data[['meter_reading', 'month', 'hour']]  
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_for_model)

# Aplicar Isolation Forest
iso_forest = IsolationForest(contamination=0.01, random_state=42)
iso_forest.fit(data_scaled)

# Predecir las anomalías
data['anomaly'] = iso_forest.predict(data_scaled)
data['anomaly'] = data['anomaly'].apply(lambda x: 1 if x == -1 else 0)

# Mostrar las anomalías detectadas
anomalies_iso = data[data['anomaly'] == 1]
print(f"Anomalías detectadas por Isolation Forest: {anomalies_iso.shape[0]}")
```

```python
# Creacion de true_anomalies
data['true_anomalies'] = ((data['meter_reading'] < lower_limit) | (data['meter_reading'] > upper_limit)).astype(int)
# Ejemplo de cálculo de métricas
precision = precision_score(data['true_anomalies'], data['anomaly'])
recall = recall_score(data['true_anomalies'], data['anomaly'])
f1 = f1_score(data['true_anomalies'], data['anomaly'])

print(f"Precisión: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
```

```python

X_train, X_test, y_train, y_test = train_test_split(data_for_model, data['true_anomalies'], test_size=0.2, random_state=42)

# Entrenar el modelo (ejemplo con Isolation Forest)
iso_forest.fit(X_train)

# Predecir en el conjunto de prueba
y_pred = iso_forest.predict(X_test)

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión en los datos de prueba: {accuracy}")
```
